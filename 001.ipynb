```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
from tqdm import tqdm
from readability import Document
import re
import os
```

```python
def get_url(url):
    response = requests.get(url)
    
    # 设置响应内容的编码为GBK
    response.encoding = 'UTF-8'

    if response.status_code == 200:
    # 使用BeautifulSoup解析HTML内容
        return response.text
    
    else:
        print("请求失败，状态码:", response.status_code)

        return None
```

```python

def get_headlines(url):
    """从人行新闻列表的页面，获得全部新闻时间、标题和链接
    
    例如:
    url = 'http://www.pbc.gov.cn/goutongjiaoliu/113456/113469/index.html'
    
    """

    # 发送HTTP请求
    response = requests.get(url)


    # 设置响应内容的编码为GBK
    response.encoding = 'UTF-8'

    result = []

    # 确保请求成功
    if response.status_code == 200:
        # 使用BeautifulSoup解析HTML内容
        soup = BeautifulSoup(response.text, 'html.parser')
        
        # 这里是一个示例，提取并打印网页的标题
        title = soup.find('title').text
        # print("网页标题:", title)


        # 找到id为"r_con"的div
        r_con_div = soup.find('div', id='r_con')
        
        # 确保找到了该div
        if r_con_div:

            td_tags = r_con_div.find_all('td')
            # print(td_tags)

            for td in td_tags:
                a_tag = td.find('a')
                if a_tag is not None:
                    #print("标题:", a_tag.text, "链接", a_tag.get('href'))
                    #print('时间:', td.find('span').text)
                    if td.find('span').text !='':
                        result.append({'time':td.find('span').text,'title':a_tag.text,'url':'http://www.pbc.gov.cn' + a_tag.get('href')})

            result = pd.DataFrame(result)

        else:
            print('未找到id为"r_con"的div元素')

        # 您可以根据需要添加更多解析代码，比如提取特定标签的内容
    else:
        print("请求失败，状态码:", response.status_code)

    return result


```

```python
result = []


for i in tqdm(range(1,357+1)):
    url = f'http://www.pbc.gov.cn/goutongjiaoliu/113456/113469/11040/index{i}.html'
    result.append(get_headlines(url))
    time.sleep(0.2)
    
df = pd.concat(result)
df.to_excel('人民银行_新闻.xlsx',index=False)
```

```python
def get_content(url):
    """从新闻页面，获得新闻本体
    例如:
    url = 'http://www.pbc.gov.cn/goutongjiaoliu/113456/113469/5194629/index.html'
    
    """
    soup = BeautifulSoup(get_url(url), 'html.parser')
    zoom_div = soup.find('div',id='zoom')
    zoom_div.find_all('p')

    doc = Document(get_url(url))
    result = doc.title() +'\n\n' + BeautifulSoup(doc.summary(), 'html.parser').get_text().strip()

    return result

print(get_content('http://www.pbc.gov.cn/goutongjiaoliu/113456/113469/5194629/index.html'))
```

```python


def sanitize_filename(file_name):
    # 移除Linux不允许的字符
    file_name = file_name.replace('/', '_').replace('\0', '_')

    # 移除Windows不允许的字符
    file_name = re.sub(r'[<>:"/\\|?*]', '_', file_name)

    # 替换Windows保留字
    reserved_words = ['CON', 'PRN', 'AUX', 'NUL', 'COM1', 'COM2', 'COM3', 'COM4', 'COM5', 'COM6', 'COM7', 'COM8', 'COM9', 'LPT1', 'LPT2', 'LPT3', 'LPT4', 'LPT5', 'LPT6', 'LPT7', 'LPT8', 'LPT9']
    base_name, extension = os.path.splitext(file_name)
    if base_name.upper() in reserved_words:
        file_name = '_' + file_name

    # 确保文件名不以空格或点结束
    file_name = file_name.rstrip('. ')

    return file_name

sanitize_filename('2023-06-08_积极践行绿色发展理念 促进30/60目标平稳实现——人民银行行长易纲在第十四.txt')
```
